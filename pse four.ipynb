### Problem 2
In this dataset, there is a column called `data_quality` which indicates which of our rows contains data that is reliable. Explore this variable a little by:
* determining the number of unreliable vs. reliable rows
* the country (or countries) with the highest number of unreliable rows
* then the country (or countries) with the highest proportion of unreliable rows (# of unreliable rows/# of total rows for each country)
We do this to understand a little more able how reliable our data is, and what some of the consequence will be if we choose to remove unreliable data.
For these tasks, you will want to use techniques shown in [Ch. 10 Data Aggregation and Group Operations](https://wesmckinney.com/book/data-aggregation). (5 pts.)

```python
#Determining the number of unreliable vs. reliable rows
unreliable_count = df['data_quality'].value_counts().get(0)
reliable_count = df['data_quality'].value_counts().get(1)

print("Number of Unreliable Rows:", unreliable_count)
print("Number of Reliable Rows:", reliable_count)

#The Country (or countries) with the highest number of unreliable rows
country_highest_unreliability=df[df['data_quality']==0]['Country'].value_counts().idxmax()
print(f'Country with highest unreliable rows:{country_highest_unreliability}')


# the country (or countries) with the highest proportion of unreliable rows (# of unreliable rows/# of total rows for each country) We do this to understand a little more able how reliable our data is, and what some of the consequence will be if we choose to remove unreliable data.
total_counts = df.groupby('Country').size()

proportions = unreliable_count / total_counts

highest_proportion_country = proportions.idxmax()
highest_proportion = proportions.max()

print("Country with the Highest Proportion of Unreliable Rows:", highest_proportion_country)
print("Highest Proportion of Unreliable Rows:", highest_proportion)
```

### Problem 5
In how many cities is it cheaper to buy beef than it is to buy chicken? Print this result and the city names using your new `City_Country` column values. (2 pts.)

```python
df['Beef_Cheaper'] = df['Beef'] < df['Chicken']

cheaper_beef_cities =df.loc[df['Beef_Cheaper'], 'City_Country'].unique()

num_cities = len(cheaper_beef_cities)

print("Number of Cities Where Beef is Cheaper Than Chicken:", num_cities)
print("City Names Where Beef is Cheaper Than Chicken:")
for city in cheaper_beef_cities:
    print(city)

```
